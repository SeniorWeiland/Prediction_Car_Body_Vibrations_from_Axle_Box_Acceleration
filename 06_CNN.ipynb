#### Model Opt

```python
from keras import initializers
p_kernel = [100] #,400]
p_units = [5]
p_layer = [2]

Train_hist = pd.DataFrame()


for layer in tqdm(p_layer):
    for kernel in tqdm(p_kernel):
        for unit in tqdm(p_units):
            NAME = "CNN4_{}layer-{}kernel-{}units".format(layer,kernel,unit)
            name = "{}lay-{}ker-{}un".format(layer,kernel,unit)

            # Define Model
            model = Sequential()
            model.add(Masking(mask_value=0.0))
            model.add(InputLayer((None,None,1), sparse=False))
            for i in range(layer):
                model.add(Conv1D(unit, kernel_size = kernel, padding='same', activation='tanh',kernel_initializer=initializers.RandomNormal(stddev=1)))
            model.add(Conv1D(1, kernel_size = 1, activation='linear',kernel_initializer=initializers.RandomNormal(stddev=2))) # default activation
                                                                                                                                                                                                             
            # Compile with MSE
            model.compile(loss='mse',
                          optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                          metrics=['mse','cosine_similarity'])

            # Define Callbacks
            tb = TensorBoard(log_dir='./logs/{}'.format(NAME))
            cp = ModelCheckpoint(NAME, save_best_only=True)
            reduce_lr = ReduceLROnPlateau(monitor='mse', factor=0.1,
                              patience=5, min_lr=0.000001)
            
            # Fit Models
            history = model.fit(X_train_arr,y_train_arr,validation_data=(X_val_arr, y_val_arr), epochs=40, callbacks=[ tb,reduce_lr], verbose=2)

            # Save History
            loss = history.history['loss']
            val_loss = history.history['val_loss']

            # Create df via dict 
            dict = {name + '_loss': loss, name + '_val_loss': val_loss } 
            hist = pd.DataFrame(dict)

            # Concat Training History for all journeys
            Train_hist = pd.concat([Train_hist, hist], axis = 1)

            
            
```
