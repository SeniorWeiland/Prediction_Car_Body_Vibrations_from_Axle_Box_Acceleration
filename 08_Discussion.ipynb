### Train Models with A-Rated Dataset

#### NN:

```python
import datetime
ct = datetime.datetime.now()
#NAME = "NN_-{}".format(str(str(ct.month)+'_'+str(ct.day)+'_'+str(ct.hour)+'_'+str(ct.minute)+'_'+str(ct.second)))
NAME = "NN_final"

nn_model = Sequential()
nn_model.add(Input(shape=(window,1)))
nn_model.add(Dense(window,'relu',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1)))
nn_model.add(Dense(window,'relu',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1)))
nn_model.add(Dense(1,'tanh',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1)))

# Configure the model 
print(NAME)
nn_model.summary()
nn_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss = 'mse',
    #loss = 'cosine_similarity',
    metrics=['mse','cosine_similarity']
)

#Define Callbacks 
tb = TensorBoard(log_dir='./logs/{}'.format(NAME))
cp = ModelCheckpoint(NAME, save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=0, min_lr=1e-8, vorbose = 1)
earlystop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=10,
    verbose=1,
    mode="min",
    baseline=None,
    restore_best_weights=False
)
```

```python
# Train Model
history_nn = nn_model.fit(
    x=X_train_mod_w,
    y=y_train_mod_w,
    batch_size=10,
    epochs=100,
    verbose=1,
    callbacks=[cp, tb, reduce_lr, earlystop ],  #, reduce_lr],
    validation_split=0.2,
    validation_data=None,
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
    validation_batch_size=None,
    validation_freq=1,
)
```

```python
# Evaluate Models for all data sets
nn_train_mod_acc = nn_model.evaluate(X_train_mod_w, y_train_mod_w)[0]
nn_train_acc = nn_model.evaluate(X_train_w, y_train_w)[0]
nn_val_acc = nn_model.evaluate(X_val_w, y_val_w)[0]
nn_test_acc = nn_model.evaluate(X_test_w, y_test_w)[0]

#Create prediction for plotting
nn_y_p = nn_model.predict(X_val_w)
nn_y_p[0].shape
```
