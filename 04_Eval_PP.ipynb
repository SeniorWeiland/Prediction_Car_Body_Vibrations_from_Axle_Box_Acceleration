#### Function Definition

```python
# Define Function for Calculating Metrics fo eachpair of Sparse Arrays
def process_data(x_sp, y_sp):
    from tslearn.metrics import dtw, dtw_path
    from scipy.stats import pearsonr
    
    DTW = []
    CosSin = []
    CrossCorr = []
    Pear = []
    
    for jrn in range(x_sp.shape[0]):

        # load journey data
        x = x_sp[jrn].data.astype(float)
        y = y_sp[jrn].data.astype(float)

        #qualize length
        if x.size != y.size:
            x_eq = eq_adc_imu_np(x,y)
        else: x_eq = x

        # DTW Score
        if len(x_eq)>50000:
            DTW.append(int(100))
        else: 
            DTW.append(dtw(x_eq, y))

        # Cosine Similarity
        CosSin.append(np.dot(x_eq,y)/(np.linalg.norm(x_eq)*np.linalg.norm(y)))

        # Cross Correlation
        xy = np.max(np.abs(np.correlate(x_eq, y, mode='same')))
        xx = np.max(np.abs(np.correlate(x_eq, x_eq, mode='same')))
        yy = np.max(np.abs(np.correlate(y, y, mode='same')))
        CrossCorr.append((2*xy)/(xx+yy))
        
        # Pearson Coefficient
        corr, pvalue = pearsonr(x_eq,y)
        Pear.append(corr)
        
        

    return DTW,Pear,CosSin, CrossCorr
```

#### Function Call

```python
# raw
DTW1_raw,Pear1_raw,CosSin1_raw,CrossCorr1_raw = process_data(s1_adc_raw, s1_imu_raw)
DTW2_raw,Pear2_raw,CosSin2_raw,CrossCorr2_raw = process_data(s2_adc_raw, s2_imu_raw)
DTW3_raw,Pear3_raw,CosSin3_raw,CrossCorr3_raw = process_data(s3_adc_raw, s3_imu_raw)
DTW4_raw,Pear4_raw,CosSin4_raw,CrossCorr4_raw = process_data(s4_adc_raw, s4_imu_raw)
DTW5_raw,Pear5_raw,CosSin5_raw,CrossCorr5_raw = process_data(s5_adc_raw, s5_imu_raw)
# filtered
DTW1_f,Pear1_f,CosSin1_f,CrossCorr1_f = process_data(s1_adc_f, s1_imu_f)
DTW2_f,Pear2_f,CosSin2_f,CrossCorr2_f = process_data(s2_adc_f, s2_imu_f)
DTW3_f,Pear3_f,CosSin3_f,CrossCorr3_f = process_data(s3_adc_f, s3_imu_f)
DTW4_f,Pear4_f,CosSin4_f,CrossCorr4_f = process_data(s4_adc_f, s4_imu_f)
DTW5_f,Pear5_f,CosSin5_f,CrossCorr5_f = process_data(s5_adc_f, s5_imu_f)
# synced
DTW1_sync,Pear1_sync,CosSin1_sync,CrossCorr1_sync = process_data(s1_adc_sync, s1_imu_sync)
DTW2_sync,Pear2_sync,CosSin2_sync,CrossCorr2_sync = process_data(s2_adc_sync, s2_imu_sync)
DTW3_sync,Pear3_sync,CosSin3_sync,CrossCorr3_sync = process_data(s3_adc_sync, s3_imu_sync)
DTW4_sync,Pear4_sync,CosSin4_sync,CrossCorr4_sync = process_data(s4_adc_sync, s4_imu_sync)
DTW5_sync,Pear5_sync,CosSin5_sync,CrossCorr5_sync = process_data(s5_adc_sync, s5_imu_sync)
# preprocessed (scaled)
DTW1,Pear1,CosSin1,CrossCorr1 = process_data(s1_adc, s1_imu)
DTW2,Pear2,CosSin2,CrossCorr2 = process_data(s2_adc, s2_imu)
DTW3,Pear3,CosSin3,CrossCorr3 = process_data(s3_adc, s3_imu)
DTW4,Pear4,CosSin4,CrossCorr4 = process_data(s4_adc, s4_imu)
DTW5,Pear5,CosSin5,CrossCorr5 = process_data(s5_adc, s5_imu)
```
