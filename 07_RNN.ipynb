#### Small Opt run

```python
#Generate Models and predictions

from keras import initializers
from keras import regularizers

p_units = [1,5,10]
p_layer = [1,2]

Train_hist = pd.DataFrame()
for layer in tqdm(p_layer):
    for unit in tqdm(p_units):
        NAME = "RNN1_{}layer-{}units".format(layer,unit)
        name = "{}lay-{}un".format(layer,unit)

        # Input Layer and Masking
        model = Sequential()
        model.add(Masking(mask_value=0.0))
        model.add(InputLayer((None,None,1), sparse=False))

        # LSTM Layers
        for i in range(layer):
            model.add(LSTM(unit, activation='tanh', return_sequences=True,kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1))) #,kernel_regularizer=regularizers.L1L2(l1=1e-5, l2=1e-4)))
        
        # Time DIstributed Dense Net for creating output
        model.add(TimeDistributed(Dense(1,activation='linear',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1))))
        model.build(input_shape=X_train_arr.shape)
        model.summary()
                                                                                                                                                                                                         
        # Compile with MSE
        model.compile(loss='mse',
                      optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                      metrics=['mse','cosine_similarity'])

        tb = TensorBoard(log_dir='./logs/{}'.format(NAME))
        cp = ModelCheckpoint(NAME, save_best_only=True)
        reduce_lr = ReduceLROnPlateau(monitor='mse', factor=0.1,
                          patience=5, min_lr=0.000001)
        
        # Fit Models
        history = model.fit(X_train_arr,y_train_arr,validation_data=(X_val_arr, y_val_arr), epochs=300, callbacks=[tb,reduce_lr], verbose=2)

        # Save History
        loss = history.history['loss']
        val_loss = history.history['val_loss']

        # Create df via dict 
        dict = {name + '_loss': loss, name + '_val_loss': val_loss } 
        hist = pd.DataFrame(dict)

        Train_hist = pd.concat([Train_hist, hist], axis = 1)
```
