#### Create Window Dataset 

```python
# Create Window Dataset
window = 100
#regular data
X_train_w, y_train_w = sparse_to_window(X_train, y_train, window)
X_val_w, y_val_w = sparse_to_window(X_val, y_val, window)
X_test_w, y_test_w = sparse_to_window(X_test, y_test, window)
#MinMaxScaled Data
# X_train_w, y_train_w = sparse_to_window(X_train_msc, y_train_msc, window)
# X_val_w, y_val_w = sparse_to_window(X_val_msc, y_val_msc, window)
# X_test_w, y_test_w = sparse_to_window(X_test_msc, y_test_msc, window)

print(X_train_w.shape)
print(X_val_w.shape)
print(X_test_w.shape)
```

#### Create Model

```python
import datetime
ct = datetime.datetime.now()
#NAME = "NN_-{}".format(str(str(ct.month)+'_'+str(ct.day)+'_'+str(ct.hour)+'_'+str(ct.minute)+'_'+str(ct.second)))
NAME = "NN_trial"

model1 = Sequential()
model1.add(Input(shape=(window,1)))
model1.add(Dense(10,'tanh',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1)))
model1.add(Dense(1,'linear',kernel_initializer=tf.keras.initializers.RandomNormal(stddev=1)))

# Configure the model 
print(NAME)
model1.summary()
model1.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss = 'mse',
    #loss = 'cosine_similarity',
    metrics=['mse','cosine_similarity']
)

#Define Callbacks 
tb = TensorBoard(log_dir='./logs/{}'.format(NAME))
cp = ModelCheckpoint(NAME, save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=0, min_lr=1e-8, vorbose = 1)

# Train Model
model1.fit(
    x=X_train_w,
    y=y_train_w,
    batch_size=100,
    epochs=5,
    verbose=1,
    callbacks=[cp, tb],  #, reduce_lr],
    validation_split=0.0,
    validation_data=(X_val_w, y_val_w),
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
    validation_batch_size=None,
    validation_freq=1,
)
```
